```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('HongKong_China_Data.csv')
data.head(5)
```

```python
data.info()
data2 = data.copy()
```

```python
#Identificar valores nulos por columna
valores_nulos = data.isnull().sum()
valores_nulos
```

Reemplazar valores nulos con el método de "mediana" en las columnas tipo "numeric (int y float)"

```python
data2["reviews_per_month"]=data2["reviews_per_month"].fillna(round(data["reviews_per_month"].median(),1))
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
data2 ["review_scores_rating"] = data2["review_scores_rating"].fillna(round(data2["review_scores_rating"].median(),1))
valores_nulos = data2.isnull().sum()
valores_nulos
```

Reemplazar valores nulos con el método de "ffill" en las columnas tipo "object, datetime y category"

```python
data2 ["host_response_time"] = data2["host_response_time"].fillna(method="ffill")
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
data2 ["host_response_rate"] = data2["host_response_rate"].fillna(method="ffill")
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
data2 ["first_review"] = data2["first_review"].fillna(method="ffill").fillna(method="ffill")
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
data2 ["bedrooms"] = data2["bedrooms"].fillna(method="ffill").fillna(method="ffill")
valores_nulos = data2.isnull().sum()
valores_nulos
```
